```python
import pandas as pd

# Replace 'your_file.csv' with the actual filename of your CSV file.
df = pd.read_csv('craigslist_vehicles.csv')

# Display the first few rows of the DataFrame to check if the data has been loaded correctly.
df.head()
```

```python
# Identify numerical and categorical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Calculate the median for numerical columns and the mode for categorical columns
median_values = df[numerical_columns].median()
mode_values = df[categorical_columns].mode().iloc[0]

# Fill in missing values with medians and modes
df[numerical_columns] = df[numerical_columns].fillna(median_values)
df[categorical_columns] = df[categorical_columns].fillna(mode_values)

# Check for any remaining missing values
missing_values = df.isnull().sum()
```
